# 웹(Naver tv) 데이터 가져오기

```python
'''라이브러리 가져오기'''
import requests                          # 웹 페이지의 데이터를 가져올 때 사용
from bs4 import BeautifulSoup as soup    # HTML코드를 parsing하여 원하는 데이터를 추출할 때 사용
```

```python
'''웹에서 데이터 가져오기'''
raw = requests.get("https://tv.naver.com/r")
```

```python
print(raw)
```

```python
print(raw.text)
```

```python
'''HTML 소스코드 parsing하기'''
parsed_data = soup(raw.text, "html.parser")
```

```python
print(parsed_data)
```

```python
'''컨테이너 수집'''
clip_titles = parsed_data.select("dt.title")
```

```python
print(clip_titles)
```

```python
'''영상별 데이터 수집1'''
clip_title_one = parsed_data.select_one("dt.title")
```

```python
print(clip_title_one)
```

```python
'''영상별 데이터 수집2'''
clip_info = parsed_data.select("div.inner")

title = clip_info[0].select_one("dt.title")
chn = clip_info[0].select_one("dd.chn")
hit = clip_info[0].select_one("span.hit")
like = clip_info[0].select_one("span.like")

print(title.text.strip())
print(chn.text.strip())
print(hit.text.strip())
print(like.text.strip())
```

```python
'''영상별 데이터 수집 반복하기'''
for clip in clip_info :
    title = clip.select_one("dt.title")
    chn = clip.select_one("dd.chn")
    hit = clip.select_one("span.hit")
    like = clip.select_one("span.like")

    print(title.text.strip())
    print( chn.text.strip())
    print( hit.text.strip())
    print( like.text.strip())
```

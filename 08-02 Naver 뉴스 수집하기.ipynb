# Naver 뉴스 수집하기

```python
'''라이브러리 가져오기'''
import requests                          # 웹 페이지의 데이터를 가져올 때 사용
from bs4 import BeautifulSoup as soup    # HTML코드를 parsing하여 원하는 데이터를 추출할 때 사용
```

```python
'''네이버에서 "코알라"로 검색한 결과 url의 데이터 가져오기'''
raw = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=코로나", headers={"User-Agent":"Mozilla/5.0"})   # header 이후부분은 웹의 크롤링 방지기능을 모질라 웹브라우저로 위장하므로써 회피하기 위함.
```

```python
print(raw)
```

```python
html = soup(raw.text, 'html.parser')
```

```python
html
```

```python
# 컨테이너 : ul.type01 > li
# 기사제목 : a._sp_each_title
# 언론사 : span._sp_each-source
```

```python
table = html.find('ul',{'class' : 'list_news'})
```

```python
table
```

```python
articles = html.select("ul > li")
```

```python
articles
```

```python
for ar in articles:
    title = ar.select_one("a._sp_each_title")
    source = ar.select_one("span._sp_each_source")
    
    print(title, source)
```
